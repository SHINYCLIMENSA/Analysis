```python
#Grafo dirigido
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
from collections import Counter
import itertools
import numpy as np
import matplotlib
```

```python
data = pd.read_csv('C:/Users/shiny/OneDrive/Desktop/RASFF 2021-2024.csv', sep=';', 
                   header=0, index_col = 0)
```

```python
countries=data[['ORIG_COUNTRY','DESTIN_COUNTRY','CONCERN_COUNTRY','CLAS_SIFICATION']]
```

```python
import networkx as nx
import pandas as pd

# Define the graphs for Total, Alerts, Border rejection, and Information
G_total = nx.DiGraph()
G_alerts = nx.DiGraph()
G_border_rejection = nx.DiGraph()
G_information_attention = nx.DiGraph()
G_information_followup = nx.DiGraph()

# Loop through the DataFrame
for index, row in countries.iterrows():
    # Convert to strings and handle NaN values
    origin = str(row['ORIG_COUNTRY']) if pd.notna(row['ORIG_COUNTRY']) else ''
    destiny = str(row['DESTIN_COUNTRY']) if pd.notna(row['DESTIN_COUNTRY']) else ''
    concern = str(row['CONCERN_COUNTRY']) if pd.notna(row['CONCERN_COUNTRY']) else ''
    case_type = str(row['CLAS_SIFICATION']).strip().lower() if pd.notna(row['CLAS_SIFICATION']) else ''

    # Split origin, destiny, and concern only if the string is not empty
    origin = origin.split(",") if origin else []
    destiny = destiny.split(",") if destiny else []
    concern = concern.split(",") if concern else []
    
    for i in origin:
        if i not in [" ", "INFOSAN", "Commission Services", "unknown origin", "European Food Safety Authority"]:
            for j in destiny:
                if j not in [" ", i, "INFOSAN", "Commission Services", "unknown origin", "European Food Safety Authority"]:
                    # Add edges to the Total graph
                    G_total.add_edge(i, j)
                    
                    # Add edges to specific graphs based on case type
                    if case_type == "alert notification":
                        G_alerts.add_edge(i, j)
                    elif case_type == "border rejection notification":
                        G_border_rejection.add_edge(i, j)
                    elif case_type == "information notification for attention":
                        G_information_attention.add_edge(i, j)
                    elif case_type == "information notification for follow-up":
                        G_information_followup.add_edge(i, j)
                    
                    # Add concern country as an edge
                    for k in concern:
                        if k not in [" ", "INFOSAN", "Commission Services", "unknown origin", "European Food Safety Authority"]:
                            G_total.add_edge(j, k)
                            
                            # Add to specific graphs based on case type
                            if case_type == "alert notification":
                                G_alerts.add_edge(j, k)
                            elif case_type == "border rejection notification":
                                G_border_rejection.add_edge(j, k)
                            elif case_type == "information notification for attention":
                                G_information_attention.add_edge(j, k)
                            elif case_type == "information notification for follow-up":
                                G_information_followup.add_edge(j, k)
                else:
                    for k in concern:
                        if k not in [" ", "INFOSAN", "Commission Services", "unknown origin", "European Food Safety Authority"]:
                            G_total.add_edge(i, k)
                            
                            # Add to specific graphs based on case type
                            if case_type == "alert notification":
                                G_alerts.add_edge(i, k)
                            elif case_type == "border rejection notification":
                                G_border_rejection.add_edge(i, k)
                            elif case_type == "information notification for attention":
                                G_information_attention.add_edge(i, k)
                            elif case_type == "information notification for follow-up":
                                G_information_followup.add_edge(i, k)

# Now you have five separate graphs: 
# G_total, G_alerts, G_border_rejection, G_information_attention, and G_information_followup
```
