```python
data = pd.read_csv('C:/Users/shiny/OneDrive/Desktop/RASFF 2021-2024.csv', sep=';', 
                   header=0, index_col = 0)
```

```python
# Initialize lists to store results
origen_not = []
rel_not = []
dest_not = []
ori_dest_not = []
ori_rel_not = []
dest_rel_not = []
route_dist = []
route_dest = []

# Iterate over each row
for index, row in data.iterrows():
    # Replace NaN with an empty string to handle missing values
    not_country = row['NOT_COUNTRY'] if pd.notna(row['NOT_COUNTRY']) else ''
    orig_country = row['ORIG_COUNTRY'] if pd.notna(row['ORIG_COUNTRY']) else ''
    count_destin = row['DESTIN_COUNTRY'] if pd.notna(row['DESTIN_COUNTRY']) else ''
    count_concern = row['CONCERN_COUNTRY'] if pd.notna(row['CONCERN_COUNTRY']) else ''
   
    # Apply conditions only if NOT_COUNTRY is valid
    if not_country:
        if not_country in orig_country and not_country in count_destin:
            ori_dest_not.append(not_country)
        if not_country in orig_country and not_country not in count_destin:
            origen_not.append(not_country)
        if not_country in count_destin and not_country not in orig_country:
            dest_not.append(not_country)
            if not_country in ['Italy', 'Belgium', 'Germany', 'Netherlands', 'France', 'Poland', 'Croatia']:
                route_dest.append([not_country, orig_country])  # Record route with orig_country
        if not_country in count_concern:
            rel_not.append(not_country)
            if not_country in ['Italy', 'Belgium', 'Germany', 'Spain', 'Netherlands', 'Croatia', 'Poland']:
                route_dist.append([not_country, orig_country])  # Record route with orig_country
        if not_country in count_concern and not_country in orig_country:
            ori_rel_not.append(not_country)
        if not_country in count_concern and not_country in count_destin:
            dest_rel_not.append(not_country)
```

```python
routes = pd.DataFrame(route_dest)
routes.columns = 'Notificante', 'Origen'
```

```python
alerts_routes = pd.concat(frames)
alerts_routes_count = alerts_routes.reset_index(drop=True)
alerts_routes_count
```

```python
alert_nodes = alerts_routes_count['Notificante'].unique().tolist()
alert_nodes. append(alerts_routes_count['Origen'].unique().tolist())
```

```python
def flatten(A):
    rt = []
    for i in A:
        if isinstance(i,list): rt.extend(flatten(i))
        else: rt.append(i)
    return rt
```

```python
alert_nodes = flatten(alert_nodes)
alert_nodes = alert_nodes
alert_nodes = list(set(alert_nodes))
```

```python
def swapPositions(list, pos1, pos2): 
      
    list[pos1], list[pos2] = list[pos2], list[pos1] 
    return list
```

```python
Germany_ix = alert_nodes.index("Germany")
alert_nodes = swapPositions(alert_nodes, Germany_ix, 0)
Belgium_ix = alert_nodes.index("Belgium")
alert_nodes = swapPositions(alert_nodes, Belgium_ix, 1)
France_ix = alert_nodes.index("France")
alert_nodes = swapPositions(alert_nodes, France_ix, 2)
Netherlands_ix = alert_nodes.index("Netherlands")
alert_nodes = swapPositions(alert_nodes, Netherlands_ix, 3)
Italy_ix = alert_nodes.index("Italy")
alert_nodes = swapPositions(alert_nodes, Italy_ix, 4)
Poland_ix = alert_nodes.index("Poland")
alert_nodes = swapPositions(alert_nodes, Poland_ix, 5)
Croatia_ix = alert_nodes.index("Croatia")
alert_nodes = swapPositions(alert_nodes,Croatia_ix, 6)
```

```python
alert_nodes = pd.DataFrame(alert_nodes)
alert_nodes.columns = ['Paises']
alert_nodes
```
